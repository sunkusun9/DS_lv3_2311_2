```python
# ì‹¤í–‰ í™˜ê²½ í™•ì¸

import pandas as pd
import numpy as np
import sklearn
import scipy
import statsmodels
import mlxtend
import sys

print(sys.version)
for i in [pd, np, sklearn, scipy, mlxtend, statsmodels]:
    print(i.__name__, i.__version__)
```

```python
df_elec = pd.read_csv('elec.csv', parse_dates=['Date', 'DateHour'])
df_info = pd.read_csv('info.csv')
df_info['Date'] = pd.to_datetime(df_info['Date'])
```

```python
df_elec1 = df_elec.pivot(
    index='DateHour', 
    columns='Minute', 
    values='Value').reset_index() # indexì— ìœ„ì¹˜í•œ DateHourë¥¼ ì»¬ëŸ¼ì— ìœ„ì¹˜ì‹œí‚µë‹ˆë‹¤
df_elec1
```

```python
holi = pd.to_datetime(["2021-01-01", "2021-02-11", "2021-02-12", "2021-03-01", "2021-05-05", "2021-05-19", "2021-08-16"]).date

# pd.Series.dt accessorë¥¼ í†µí•´ íŒŒìƒ ë³€ìˆ˜ë“¤ì„ ë§Œë“­ë‹ˆë‹¤.
df_elec1 = df_elec1.assign(
    DayName = df_elec1['DateHour'].dt.weekday, 
    Hour = df_elec1['DateHour'].dt.hour,
    AM = (df_elec1['DateHour'].dt.hour >= 12).astype('int'),
    Weekend_yn = df_elec1['DateHour'].dt.weekday.isin([5, 6]).astype('int'),
    Holiday_yn = df_elec1['DateHour'].dt.date.isin(holi).astype('int'), # df_elec1['DateHour'].dt.dateëŠ” pd.Series ì…ë‹ˆë‹¤.
    Avg = df_elec1.mean(axis=1),
    TotalHour = df_elec1.sum(axis=1),
)
```

```python
df_info1 = df_info.fillna(0)
```

```python
df_basetable1 = df_elec1.merge(
    df_info1, 
    left_on='DateHour', 
    right_on='Date', how='inner'
).drop(columns='Date') # DateHourê°€ ì»¬ëŸ¼ì— ìˆìœ¼ë¯€ë¡œ DateHour, Date ëª¨ë‘ ì¡´ì¬í•˜ì—¬, DateëŠ” ì‚­ì œí–ˆìŠµë‹ˆë‹¤.
display(df_basetable1.head())
df_basetable1.shape
```

```python
# shiftë¥¼ ì´ìš©í•©ë‹ˆë‹¤.
df_prob3 = df_basetable1.sort_values('DateHour')
df_prob3['target'] = df_prob3['TotalHour'].shift(-24)
df_prob4 = pd.concat(
    [df_prob3] + [
        df_prob3['TotalHour'].shift(24 * i).rename('lag_{}'.format(i)) for i in range(1, 7)
    ], axis=1
).dropna()
df_prob4.head()
```

```python
# shiftë¥¼ ì‚¬ìš©í•˜ì§€ ì•Šê³  targetë¥¼ êµ¬í•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. (ê°’ì„ êµ¬í•˜ê¸°ë§Œ í•˜ê³ , ë”°ë¡œ prob4ë¡œ ë§Œë“¤ì§€ëŠ” ì•Šì•˜ìŠµë‹ˆë‹¤.)
df_basetable1.join(
    df_basetable1[['DateHour','TotalHour']].assign(
        DateHour=lambda x: x['DateHour'] - pd.Timedelta(days=1)
    ).set_index('DateHour').rename(columns={'TotalHour': 'target'}),
    on='DateHour', how='inner'
)
```

```python
# ë¬¸ì œì—ì„œ ì œì‹œí•œ ë°ì´í„°ì…‹ì„ ë§Œë“­ë‹ˆë‹¤.
df_prob6_train = df_prob4.loc[df_prob4['DateHour'] < '2021-08-14']
df_prob6_test = df_prob4.loc[df_prob4['DateHour'] >= '2021-08-14']
# ì •ë‹µ targetì€ ë–¼ì–´ëƒ…ë‹ˆë‹¤.
s_kaggle_ans =  df_prob6_test.pop('target')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_absolute_error

# ë¬¸ì œì—ì„œ ì œì‹œí•œ ë™ì¼í•œ ê¸°ì¤€ìœ¼ë¡œ, train/testë¥¼ ë‚˜ëˆ•ë‹ˆë‹¤.
df_train = df_prob6_train.loc[df_prob6_train['DateHour'] < '2021-07-14']
df_test = df_prob6_train.loc[df_prob6_train['DateHour'] >= '2021-07-14']

# testì— ìˆëŠ” ë³€ìˆ˜ëŠ” ëª¨ë‘ ì…ë ¥ë³€ìˆ˜ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.
X = df_prob6_test.columns.tolist()

# ëª¨ë¸ì„ í‰ê°€í•˜ê¸° ìœ„í•œ í•¨ìˆ˜ì…ë‹ˆë‹¤.
def eval_model(model):
    # ê²€ì¦ìš© trainìœ¼ë¡œ í•™ìŠµí•©ë‹ˆë‹¤.
    model.fit(df_train[X], df_train['target'])
    # Trainì— ëŒ€í•œ ì„±ëŠ¥ / Test ì— ëŒ€í•œ ì„±ëŠ¥ì„ ë°˜í™˜í•©ë‹ˆë‹¤.
    return (
        mean_absolute_error(df_train['target'], model.predict(df_train[X])),
        mean_absolute_error(df_test['target'], model.predict(df_test[X]))
    )

# ëª¨ë¸ì„ ì„ íƒí•˜ê¸° ìœ„í•¨ í•¨ìˆ˜ì…ë‹ˆë‹¤. 
# ì œì¶œ íŒŒì¼ì„ ë§Œë“¤ê³ ,
# ìê°€ ì±„ì ì„ í•´ë³´ê¸° ìœ„í•œ ì˜ˆì¸¡ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
def choose_model(model):
    # ì£¼ì–´ì§„ ì „ì²´ í•™ìŠµì…‹ìœ¼ë¡œ í•™ìŠµí•©ë‹ˆë‹¤.
    model.fit(df_prob6_train[X], df_prob6_train['target'])
    # ì£¼ì–´ì§„ í…ŒìŠ¤íŠ¸ì…‹ì— ëŒ€í•œ ì˜ˆì¸¡ì„ í•©ë‹ˆë‹¤.
    prd = model.predict(df_prob6_test[X])
    pd.DataFrame({
        'DateHour': df_prob6_test['DateHour'],
        'target': prd
    }).to_csv('answer6.csv', index=None)
    return prd
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
# SVRì„ Baseline ëª¨ë¸ë¡œ ë§Œë“¤ì–´ ë´…ë‹ˆë‹¤.
# SVR ëª¨ë¸ì„ ë§Œë“¤ì–´ ì„±ëŠ¥ í‰ê°€ë¥¼ í•´ë´…ë‹ˆë‹¤.
ct = ColumnTransformer([
    ('std', StandardScaler(), ['TotalHour'] + ['lag_{}'.format(i) for i in range(1, 7)])
])
reg_svr = make_pipeline(
    ct,
    SVR(kernel='rbf', gamma=0.1, C=10)
)
eval_model(reg_svr)
```

```python
# ì¼ë‹¨ì€ Baseline ëª¨ë¸ì— ëŒ€í•œ ê²°ê³¼ë¥¼ ê¸°ë¡í•´ë‘¡ë‹ˆë‹¤.
prd = choose_model(reg_svr)
```

```python
# ìê°€ ì±„ì ì„ í•´ë´…ë‹ˆë‹¤.
mean_absolute_error(s_kaggle_ans, prd)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Baseline ëª¨ë¸ì—, DayNameì˜ ê°€ë³€ìˆ˜ë¥¼ ì¶”ê°€í•´ë´…ë‹ˆë‹¤.
ct = ColumnTransformer([
    ('std', StandardScaler(), ['TotalHour'] + ['lag_{}'.format(i) for i in range(1, 7)]),
    ('ohe', OneHotEncoder(drop='first', categories='auto'), ['DayName'])
])
reg_svr = make_pipeline(
    ct,
    SVR(kernel='rbf', gamma=0.1, C=10)
)
eval_model(reg_svr)
```

```python
# Baseline ë³´ë‹¤ëŠ” ì¢‹ìŠµë‹ˆë‹¤. 
# ì„ íƒí•©ë‹ˆë‹¤.
prd = choose_model(reg_svr)
# ìê°€ ì±„ì í•©ë‹ˆë‹¤.
mean_absolute_error(s_kaggle_ans, prd)
```

```python
from sklearn.decomposition import PCA

ct = ColumnTransformer([
    ('std_pca', make_pipeline(StandardScaler(), PCA(n_components=0.9)), ['TotalHour'] + ['lag_{}'.format(i) for i in range(1, 7)]),
    ('ohe', OneHotEncoder(drop='first', categories='auto'), ['DayName'])
])
reg_svr2 = make_pipeline(
    ct,
    SVR(kernel='rbf', gamma=0.1, C=10)
)
eval_model(reg_svr2)
```

```python
# ê²€ì¦ Train ì„±ëŠ¥ì€ ë–¨ì–´ ì¡Œì§€ë§Œ, ê²€ì¦ Test ì„±ëŠ¥ì€ ì˜¬ë¼ê°”ìŠµë‹ˆë‹¤.
# ì„ íƒí•©ë‹ˆë‹¤.
prd = choose_model(reg_svr2)
# ìê°€ ì±„ì í•©ë‹ˆë‹¤.
mean_absolute_error(s_kaggle_ans, prd)
```

```python
from sklearn.ensemble import GradientBoostingRegressor # ê²½ì‚¬ ë¶€ìŠ¤íŒ…(Gradient Boosting) íšŒê·€ëª¨ë¸ì„ ì‚¬ìš©í•œë‹¤.
from sklearn.metrics import mean_absolute_error # ğ‘€ğ´ğ¸

X_gb = "15ë¶„, 30ë¶„, 45ë¶„, 60ë¶„, DayName, Hour, AM, Weekend_yn, Holiday_yn, Avg, TotalHour,ìƒì‚°ëŸ‰, ê¸°ì˜¨, í’ì†, ìŠµë„, ê°•ìˆ˜ëŸ‰, ì „ê¸°ìš”ê¸ˆ(ê³„ì ˆ), ê³µì¥ì¸ì›, ì¸ê±´ë¹„"
X_gb = [i.strip() for i in X_gb.split(',')]

ct = ColumnTransformer([
    ('pt', 'passthrough', X_gb)
])

reg_gb = make_pipeline(
    ct,
    GradientBoostingRegressor(
        loss='huber', # ì†ì‹¤í•¨ìˆ˜: huber
        n_estimators=75, # íŠ¸ë¦¬ì˜ ìˆ˜: 75
        max_depth=7, # íŠ¸ë¦¬ì˜ ìµœëŒ€ ê¹Šì´: 7
        min_samples_leaf=4, # ë¦¬í”„ ë…¸ë“œ(Leaf Node)ì˜ ìµœì†Œ ìƒ˜í”Œ ìˆ˜: 4
        random_state=123 # ëœë¤ ì‹œë“œ: 123
    )
)
eval_model(reg_gb)
```

```python
# ë‘ ë²ˆì§¸ ëª¨ë¸ë³´ë‹¤ëŠ” ì¢‹ìŠµë‹ˆë‹¤.
# ì±„íƒí•©ë‹ˆë‹¤.
prd = choose_model(reg_gb)
# ì±„ì í•©ë‹ˆë‹¤.
mean_absolute_error(s_kaggle_ans, prd)
```

```python
# RandomForestë¡œ ëª¨ë¸ì„ ë§Œë“¤ì–´ ë´…ë‹ˆë‹¤.
from sklearn.ensemble import RandomForestRegressor

X_rf = "15ë¶„, 30ë¶„, 45ë¶„, 60ë¶„, DayName, Hour, AM, Weekend_yn, Holiday_yn, Avg, TotalHour, ìƒì‚°ëŸ‰, ê¸°ì˜¨, í’ì†, ìŠµë„, ê°•ìˆ˜ëŸ‰, ì „ê¸°ìš”ê¸ˆ(ê³„ì ˆ), ê³µì¥ì¸ì›, ì¸ê±´ë¹„"
X_rf = [i.strip() for i in  X_rf.split(',')]
ct = ColumnTransformer([
    ('pt', 'passthrough', X_rf)
])
reg_rf = make_pipeline(
    ct, 
    RandomForestRegressor(
        n_estimators=200, # íŠ¸ë¦¬ì˜ ìˆ˜: 200
        max_depth=9,# íŠ¸ë¦¬ì˜ ìµœëŒ€ ê¹Šì´: 9
        min_samples_leaf=4, # ë¦¬í”„ ë…¸ë“œ(Leaf Node)ì˜ ìµœì†Œ ìƒ˜í”Œ ìˆ˜: 4
        random_state=123, # ëœë¤ ì‹œë“œ: 123
    )
)
eval_model(reg_rf)
```

```python
# GradientBoostingRegressorë³´ë‹¤ ì¢‹ìŠµë‹ˆë‹¤.
# ì±„íƒí•©ë‹ˆë‹¤.
prd = choose_model(reg_rf)
# ì±„ì í•©ë‹ˆë‹¤.
mean_absolute_error(s_kaggle_ans, prd)
```

```python
# gb, rf, svr ì„¸ ê°œì˜ ëª¨ë¸ì„ ì•™ìƒë¸”í•´ë´…ë‹ˆë‹¤.
# rfë³´ë‹¤ ì„±ëŠ¥ì´ ì¢‹ì§€ ì•ŠìŠµë‹ˆë‹¤. ë²„ë¦½ë‹ˆë‹¤.
from sklearn.ensemble import VotingRegressor
reg_vt = VotingRegressor([
    ('svr2', reg_svr2),
    ('gb', reg_gb),
    ('rf', reg_rf),
])
eval_model(reg_vt)
```

```python
# ì•ˆ ì¢‹ìŠµë‹ˆë‹¤.
# ì±„íƒí•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.
# ì±„ì ë§Œ í•´ë´…ë‹ˆë‹¤.
reg_vt.fit(df_prob6_train[X], df_prob6_train['target'])
prd = reg_vt.predict(df_prob6_test[X])
mean_absolute_error(s_kaggle_ans, prd)
```
